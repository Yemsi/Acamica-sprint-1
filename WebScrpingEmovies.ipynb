# 3. Extracción de datos hoja visible

```python
# Click on the button to load more data
try:
    button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//button[@type="submit" and contains(@class, "form__submit button__link")]'))
    )
    button.click()
    print("Button clicked successfully!")

    # Wait for the page to load after the button click
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "course__inner")) # Assuming 'course__inner' appears after the page loads
    )
    print("Page loaded successfully after button click!")

except Exception as e:
    print("Error clicking the button or waiting for page load:", str(e))
```

```python
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize an empty list to store the extracted data
data = []

# Find all the 'course__inner' div elements
course_elements = driver.find_elements(By.CLASS_NAME, "course__inner")

# Loop through each course element and extract the required information
for course_element in course_elements:
    asignatura = None
    universidad = None
    programa = None
    enlace = None

    try:
        # Extract Asignatura (h3 value)
        h3_element = course_element.find_element(By.TAG_NAME, "h3")
        asignatura = h3_element.text.strip()
    except:
        pass # Handle cases where h3 is not found

    try:
        # Extract Universidad (strong inside sup with "IES / HEI :")
        # This assumes the structure is details__item -> sup -> strong
        universidad_element = course_element.find_element(By.XPATH, ".//div[contains(@class, 'details__item')]//sup[contains(text(), 'IES / HEI :')]/following-sibling::strong")
        universidad = universidad_element.text.strip()
    except:
        pass # Handle cases where Universidad is not found

    try:
        # Extract Programa (strong inside sup with "Programa académico / Academic Program :")
        # This assumes the structure is details__item -> sup -> strong
        programa_element = course_element.find_element(By.XPATH, ".//div[contains(@class, 'details__item')]//sup[contains(text(), 'Programa académico / Academic Program :')]/following-sibling::strong")
        programa = programa_element.text.strip()
    except:
        pass # Handle cases where Programa is not found

    try:
        # Extract Enlace (a tag inside div with class "block__button")
        block_button_div = course_element.find_element(By.CLASS_NAME, "block__button")
        enlace_element = block_button_div.find_element(By.TAG_NAME, "a")
        enlace = enlace_element.get_attribute("href")
    except:
        pass # Handle cases where block__button div or a element is not found


    # Append the extracted data to the list
    data.append({
        "Asignatura": asignatura,
        "Universidad": universidad,
        "Programa": programa,
        "Enlace": enlace
    })

# Create a pandas DataFrame
df_extracted_data = pd.DataFrame(data)

# Display the DataFrame
display(df_extracted_data)
```
